## WCSS
El método del codo es una técnica utilizada para encontrar el número óptimo de clusters en un conjunto de datos. Ayuda a identificar el punto donde el aumento en la cantidad de clusters ya no reduce significativamente la variabilidad dentro de cada cluster.
En primer lugar, ejecutamos el algoritmo K-means con diferentes valores de k (número de clusters), usualmente en un rango específico. Luego, calculamos el WCSS para cada valor de k. Esto se hace sumando las distancias al cuadrado de cada punto a su centroide y sumando este valor para todos los clusters.
Graficamos k contra WCSS. Observaremos que a medida que k aumenta, el WCSS tiende a disminuir, ya que hay más clusters y por lo tanto los puntos tienden a estar más cerca de sus respectivos centroides. Sin embargo, llegará un punto donde el descenso en WCSS se nivela. Este es el punto donde se forma un "codo" en el gráfico. El número de clusters en este punto es generalmente considerado como el número óptimo de clusters para el conjunto de datos dado.

```python
# importamos la clase KMeans dentro de sklearn
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

data_numericas = scaler.fit_transform(data_numericas)

wcss = []

for i in range(1, 10):
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=0)
    kmeans.fit(data_numericas)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 10), wcss)
plt.title('El método del codo')
plt.xlabel('Número de clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
from sklearn.cluster import KMeans

# Elegimos el número de clusters
num_clusters = 2

data_numericas = pd.DataFrame(data_numericas)
data_numericas.columns = vars

# Creamos el modelo KMeans
kmeans = KMeans(n_clusters=num_clusters)

# Entrenamos el modelo con los datos
kmeans.fit(data_numericas)

# Obtenemos las etiquetas de los clusters para cada punto de datos
etiquetas = kmeans.labels_

# Obtenemos los centroides de los clusters
centroides = kmeans.cluster_centers_

# Añadimos las etiquetas de los clusters a los datos
data_numericas["Etiqueta"] = etiquetas

# Mostramos los datos con las etiquetas de los clusters
data_numericas.head()
```

```python
# Scatterplot de Ingreso Mensual vs Valor propiedad
sns.scatterplot(data=data_numericas, x="Ingreso Mensual", y="Valor propiedad", hue="Etiqueta")
plt.title("Scatterplot de Ingreso Mensual VS Valor propiedad")
plt.show()

# Scatterplot de Ingreso Mensual vs Cantidad dormitorios
sns.scatterplot(data=data_numericas, x="Ingreso Mensual", y="Cantidad dormitorios", hue="Etiqueta")
plt.title("Scatterplot de Ingreso Mensual VS Cantidad dormitorios")
plt.show()

# Scatterplot de Ingreso Mensual vs Sup. En m2
sns.scatterplot(data=data_numericas, x="Ingreso Mensual", y="Sup. En m2", hue="Etiqueta")
plt.title("Scatterplot de Ingreso Mensual VS Sup. En m2")
plt.show()

# Scatterplot de Valor propiedad vs Cantidad dormitorios
sns.scatterplot(data=data_numericas, x="Valor propiedad", y="Cantidad dormitorios", hue="Etiqueta")
plt.title("Scatterplot de Valor propiedad VS Cantidad dormitorios")
plt.show()

# Scatterplot de Valor propiedad vs Sup. En m2
sns.scatterplot(data=data_numericas, x="Valor propiedad", y="Sup. En m2", hue="Etiqueta")
plt.title("Scatterplot de Valor propiedad VS Sup. En m2")
plt.show()

# Scatterplot de Cantidad dormitorios vs Sup. En m2
sns.scatterplot(data=data_numericas, x="Cantidad dormitorios", y="Sup. En m2", hue="Etiqueta")
plt.title("Scatterplot de Cantidad dormitorios VS Sup. En m2")
plt.show()
```
